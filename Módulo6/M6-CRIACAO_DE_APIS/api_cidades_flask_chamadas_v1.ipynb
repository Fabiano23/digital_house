### Vamos consultar o serviço na raiz do servidor web local.
Podemos também verificar isto pelo browser em seu computador

```python
# rota do serviço a ser utlizado (ver execução do cidades.py no terminal)
rota = "http://127.0.0.1:5000/"
```

```python
# utilização da biblioteca requests com o comando GET e apresentação do código de retorno
resultado = requests.get(rota)
print(resultado)
```

### Inclusão de uma nova cidade, com o método POST através de um parâmetro em 'json'

```python
rota = "http://127.0.0.1:5000/APICidades"
```

```python
params = {"nome": "Salvador"}
inclusao =  requests.post(rota, json=params)
print(resultado)
```

```python
inclusao.json()
```
