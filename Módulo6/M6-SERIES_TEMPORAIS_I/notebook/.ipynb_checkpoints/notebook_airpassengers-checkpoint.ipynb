```python
#Rolling statistics
rolmean = ts.rolling(window=12).mean()
rolstd = ts.rolling(window=12).std()
plt.figure(figsize=(18,8))   
orig = plt.plot(ts, label='Original')
mean = plt.plot(rolmean, label='Rolling Mean', c='r')
std = plt.plot(rolstd, label = 'Rolling Std', c='k')
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.title('Rolling Mean & Standard Deviation')
plt.legend()
plt.show()

# ADF Test
result = adfuller(ts, autolag='AIC')

print('Augmented Dickey Fuller test (ADH Test)')
print('='*40)
print(f'ADF Statistic: {result[0]:>24}')
print('-'*40)
print(f'p-value:       {result[1]:>24}')
print('-'*40)
print('Critial Values:')
print('-'*40)
for key, value in result[4].items():
    print(f'{key:>4} :{value:>33}')
print('='*40)
```

```python
#Rolling statistics
ts_log = np.log(ts)

rolmean_log = ts_log.rolling(window=12).mean()
rolstd_log = ts_log.rolling(window=12).std()

plt.figure(figsize=(18,8))   
orig = plt.plot(ts_log, label='Original')
mean = plt.plot(rolmean_log, label='Rolling Mean', c='r')
std = plt.plot(rolstd_log, label = 'Rolling Std', c='k')
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.title('Rolling Mean & Standard Deviation')
plt.legend()
plt.show()

# ADF Test
result = adfuller(ts_log.dropna(), autolag='AIC')

print('Augmented Dickey Fuller test (ADH Test)')
print('='*40)
print(f'ADF Statistic: {result[0]:>24}')
print('-'*40)
print(f'p-value:       {result[1]:>24}')
print('-'*40)
print('Critial Values:')
print('-'*40)
for key, value in result[4].items():
    print(f'{key:>4} :{value:>33}')
print('='*40)
```

```python
#Rolling statistics
ts_log_ma = ts_log - rolmean_log

rolmean_log_ma = ts_log_ma.rolling(window=12).mean()
rolstd_log_ma = ts_log_ma.rolling(window=12).std()

plt.figure(figsize=(18,8))   
orig = plt.plot(ts_log_ma, label='Original')
mean = plt.plot(rolmean_log_ma, label='Rolling Mean', c='r')
std = plt.plot(rolstd_log_ma, label = 'Rolling Std', c='k')
plt.xlabel("Date")
plt.ylabel("Mean Temperature")
plt.title('Rolling Mean & Standard Deviation')
plt.legend()
plt.show()

# ADF Test
result = adfuller(ts_log_ma.dropna(), autolag='AIC')

print('Augmented Dickey Fuller test (ADH Test)')
print('='*40)
print(f'ADF Statistic: {result[0]:>24}')
print('-'*40)
print(f'p-value:       {result[1]:>24}')
print('-'*40)
print('Critial Values:')
print('-'*40)
for key, value in result[4].items():
    print(f'{key:>4} :{value:>33}')
print('='*40)
```

```python
X = ts_log_ma.dropna().values

size = int(len(X) * .8)
train, test = [_ for _ in X[0:size]], [_ for _ in X[size:len(X)]]
```

```python
predict = []
model_type = {'model_name': 'arima', 'params': [(1, 0, 1)]}
for t in range(len(test)):
    if model_type['model_name'].lower() == 'arima':
        model = ARIMA(train, order=model_type['params'][0])    
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predict.append(yhat)
    obs = test[t]
    train.append(obs)
    print(f'predicted= {round(yhat[0], 5):10},  expected= {obs}')
    
error = mean_squared_error(test, predict)
print('-'*32, f'\nTest MSE: {round(error, 5)}')
```

```python
X = ts_log_ma.dropna().values

size = int(len(X) * .8)
train, test = [_ for _ in X[0:size]], [_ for _ in X[size:len(X)]]
```

```python
predictions = []
for t in range(len(test)):
    
    model = sm.tsa.statespace.SARIMAX(train, 
                                      order=(1, 0, 0), 
                                      seasonal_order=(1, 0, 1, 12), 
                                      initialization='approximate_diffuse')
    
    model_fit = model.fit(disp=0)
    
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    train.append(obs)
    
    print(f'predicted= {round(yhat, 1)},  expected= {obs}')
    
error = mean_squared_error(test, predictions)

print('-'*32, f'\nTest MSE: {round(error, 8)}')
```
