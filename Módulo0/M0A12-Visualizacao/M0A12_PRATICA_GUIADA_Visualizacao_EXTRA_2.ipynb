# PRÁTICA GUIADA 2: Visualização com Seaborn

## Conjunto de dados Iris

Esse conjunto de dados contém informações sobre a dimensão de pétalas e sépalas de três espécies de íris

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Ler o conjunto de dados
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.columns
```

* Plotar as variáveis numéricas e seus relacionamentos
* Atribuir cores de acordo com as categorias de ‘species' (hue = matiz)

```python
sns.pairplot(iris, hue='species', height=2.5);
```

* Histogramas anteriores sugerem certas sobreposições ...
* Testar com a função de densidade

```python
sns.pairplot(iris, hue="species", palette="colorblind", diag_kind="kde", height=2.5);
```

* Escolher duas variáveis de interesse
* Identificar espécies pela cor e forma do marcador

```python
FG = sns.FacetGrid(iris, hue="species", palette="Set2", height=5, 
hue_kws={"marker": ["^", "v", "*"]})
FG.map(plt.scatter, "sepal_length", "petal_length", s=100, linewidth=.5, edgecolor="white")
FG.add_legend();
```

[Aqui](http://gree2.github.io/python/2015/05/05/python-seaborn-tutorial-controlling-figure-aesthetics) há um tutorial de Seaborn para ajustar a apresentação das visualizações.

[Aqui](https://chrisalbon.com/python/seaborn_color_palettes.html) e [aqui](http://seaborn.pydata.org/tutorial/color_palettes.html) pode ser encontrada uma descrição das paletas de cores de Seaborn.

```python
sns.color_palette("husl", 8)


FG = sns.FacetGrid(iris, hue="species", palette="Set2", height=6, 
hue_kws={"marker": ["^", "v", "*"]}, aspect= 1.7)

FG.map(plt.scatter, "sepal_length", "petal_length", s=100, linewidth=.5, edgecolor="white")


FG.add_legend();
```
