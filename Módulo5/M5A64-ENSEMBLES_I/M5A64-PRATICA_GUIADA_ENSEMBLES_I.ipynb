# Ensembles - Bagging and Forests

<img src="ensembles_init.png" align="center" width="50%">

#### Até esse momento vimos como montar modelos isolados para tarefas de classificação e clusterização. Cada modelo, por seu funcionamento e capacidade, acaba tendo seus prós e contras: alguns focam em certas métricas, outros encaram o `trade-off` de viés e variância de certo modo. Dessa forma, escolher um algoritmo acarreta diretamente no comportamento e `trade-offs` do seu modelo. Mas e se não usássemos apenas um, mas [vários modelos](https://towardsdatascience.com/ensemble-methods-in-machine-learning-what-are-they-and-why-use-them-68ec3f9fef5f) que se completam. Esses são os [`Ensembles`](https://towardsdatascience.com/ensemble-models-5a62d4f4cb0c) e nos nossos próximos encontros veremos como encarar problemas utilizando técnicas de [`Ensemble`](https://towardsdatascience.com/what-are-ensemble-methods-in-machine-learning-cac1d17ed349), que nada mais são do que juntar modelos diversos para termos uma melhor modelagem.

#### O [`Ensemble Learning`](https://machinelearningmastery.com/why-use-ensemble-learning/) ajuda a melhorar os resultados do [`Machine Learning`](https://medium.com/dados-e-saude/machine-learning-e-ensembles-780f3a8aa36d) combinando vários modelos. Essa abordagem permite a produção de melhor desempenho preditivo em comparação com um único modelo. É por isso que os métodos de [`ensemble`](https://towardsdatascience.com/ensemble-methods-in-machine-learning-1f02c9bcb7fe) são os primeiros em muitas competições de aprendizado de máquina de prestígio, como a competição [`Netflix`](https://www.kaggle.com/netflix-inc/netflix-prize-data), [`KDD`](https://www.kaggle.com/c/kddbr-2020) no [`Kaggle`]().

#### Os métodos de [`Ensemble`](https://towardsdatascience.com/tagged/ensemble-learning) são [meta-algoritmos](https://medium.com/ml-research-lab/stacking-ensemble-meta-algorithms-for-improve-predictions-f4b4cf3b9237) que combinam várias técnicas de aprendizado de máquina em um modelo preditivo para diminuir a variância ([`Bagging`](https://towardsdatascience.com/ensemble-learning-bagging-boosting-3098079e5422)), o viés ([`Boosting`](https://www.kdnuggets.com/2017/11/difference-bagging-boosting.html)) ou melhorar as predisões ([`Stacking`](https://blogs.sas.com/content/subconsciousmusings/2017/05/18/stacked-ensemble-models-win-data-science-competitions/#:~:text=Model%20stacking%20is%20an%20efficient,a%20new%20set%20of%20predictions.)).


<img src="baggingBoosting.jpeg" align="center" width="50%">

#### Os métodos de *Ensembles* podem ser divididos em dois grupos:


1. [`Sequential Ensemble Methods`](https://corporatefinanceinstitute.com/resources/knowledge/other/ensemble-methods/#:~:text=Sequential%20ensemble%20techniques%20generate%20base,weights%20to%20previously%20misrepresented%20learners.): métodos de conjunto sequencial em que os [`base learners`](https://towardsdatascience.com/introduction-to-ensemble-learning-27af286ac470) são gerados sequencialmente (exemplos são os métodos de [`Boosting`](https://towardsdatascience.com/simple-guide-for-ensemble-learning-methods-d87cc68705a2)). A motivação básica dos métodos seqüenciais é explorar a dependência entre os `base learners`. O desempenho geral pode ser aprimorado pesando exemplos previamente rotulados incorretamente com maior peso.


2. [`Parallel Ensemble Methods`](https://towardsdatascience.com/ensemble-methods-bagging-boosting-and-stacking-c9214a10a205): métodos de conjunto paralelo em que os `base learners` são gerados em paralelo (exemplos são os métodos de `Bagging`, [`Stacking`, `Blending`, `Voting`](https://towardsdatascience.com/ensemble-learning-stacking-blending-voting-b37737c4f483)). A motivação básica dos métodos paralelos é explorar a independência entre os `base learners`, pois o erro pode ser reduzido drasticamente pela média.


#### A maioria dos métodos de [`Ensemble`](https://towardsdatascience.com/advanced-ensemble-learning-techniques-bf755e38cbfb) usa um algoritmo de aprendizado de base única para produzir base learners homogêneos, ou seja, alunos do mesmo tipo, levando a [`Ensembles`](https://medium.com/@dalmas.otieno/introduction-to-ensemble-methods-aca988f25fcb) homogêneos.

#### Existem também alguns métodos que usam `base learners` heterogêneos, ou seja, `learners` de diferentes tipos, levando a [`Ensembles`](https://www.datasciencecentral.com/profiles/blogs/multi-stage-heterogeneous-ensemble-meta-learning-with-hands-off) heterogêneos. Para que os métodos de [`Ensemble`](https://www.kdnuggets.com/2019/01/ensemble-learning-5-main-approaches.html) sejam mais precisos do que qualquer um de seus membros individuais, os `base learnenrs` precisam ser o mais precisos possível e o mais diversificados possível.

## Vantagens ou benefícios dos métodos de `ensemble`.

#### Os métodos de [`ensemble`](https://blogs.perficient.com/2019/11/07/predictive-model-ensembles-pros-and-cons/) são usados em quase todos os [`hackathons`](https://en.wikipedia.org/wiki/Hackathon#:~:text=A%20hackathon%20(also%20known%20as,managers%2C%20domain%20experts%2C%20and%20others) de ML para aprimorar as habilidades de previsão dos modelos. Vamos dar uma olhada nas vantagens de usar métodos de conjunto:

1. Resultados de previsão mais precisos - Podemos comparar o funcionamento dos métodos de `ensemble` à diversificação de nossas carteiras financeiras. É aconselhável manter um portfólio misto entre dívida e patrimônio para reduzir a variabilidade e, portanto, minimizar o risco. Da mesma forma, o conjunto de modelos oferecerá melhor desempenho nos cenários de caso de teste (dados não vistos) em comparação com os modelos individuais na maioria dos casos.

2. Modelo estável e mais robusto - O resultado agregado de vários modelos tem sempre menos ruído que os modelos individuais. Isso leva à estabilidade e robustez do modelo.

3. Os modelos de conjunto podem ser usados para capturar os relacionamentos lineares e não lineares nos dados. Isso pode ser realizado usando dois modelos diferentes e formando um conjunto dos dois.

## Desvantagens dos métodos de ensemble

1. Redução na capacidade de interpretação do modelo - O uso de métodos de conjunto reduz a capacidade de interpretação do modelo devido ao aumento da complexidade e torna muito difícil obter `insights` cruciais sobre os negócios no final.

2. O tempo de computação e design é alto - não é bom para aplicativos em tempo real.

3. A seleção de modelos para criar um conjunto é uma arte realmente difícil de dominar.

## Majority Voting / Hard Voting


#### Além do voto majoritário simples (votação forçada), conforme descrito na seção anterior, podemos calcular um voto majoritário associando ao classificador $C_{j}$:

$$ 
\hat{y} = mode [ C_1(x), C_2(x), ..., C_n(x) ]  
$$

#### Supondo que combinemos três classificadores que classificam uma amostra de treinamento da seguinte maneira:

    classificador 1 -> classe 0
    classificador 2 -> classe 0
    classificador 3 -> classe 1

$$ 
\hat{y} = mode [ 0, 0, 1 ] = 0  
$$

#### Por maioria de votos, classificaríamos a amostra como "classe 0".

## Weighted Majority Vote

#### Além do voto majoritário simples (votação forçada), conforme descrito na seção anterior, podemos calcular um voto majoritário ponderado associando um peso $w_{j}$ ao classificador $C_{j}$.


$$ 
\hat{y} = argmax_{i}\bigg(\sum_{j = 1}^{m} \omega_{j} X_{a} (C_{j}(x) = i)\bigg),
$$

#### onde $X_a$ é a função característica $[C_j (x) = i \in A]$ e $A$ é o conjunto de rótulos de classe exclusivos.

#### Continuando com o exemplo da seção anterior:

    classificador 1 -> classe 0
    classificador 2 -> classe 0
    classificador 3 -> classe 1

#### Podemos atribuir os pesos $\omega = \{0.2, 0.2, 0.6\}$ produziria uma previsão $\hat{y} = 1$:

$$ \hat{y} = argmax_i [ 0.2  i_{0} + 0.2  i_{0} + 0.6  i_{1} ] = 1 $$

## Soft Voting

#### Na votação `soft`, prevemos os rótulos das classes com base nas probabilidades previstas $p$ para o classificador - essa abordagem é recomendada apenas se os classificadores estiverem bem calibrados.

$$ 
\hat{y} = argmax_{i} \sum_{j=1}^{m} \omega_{i} p_{ij}  
$$

#### onde $\omega_{j}$ é o peso que pode ser atribuído ao $j$-ésimo classificador.

#### Supondo que o exemplo na seção anterior fosse uma tarefa de classificação binária com rótulos de classe $i \in \{0, 1\}$, nosso conjunto poderia fazer a seguinte previsão:

    C1(x) -> [0.9, 0.1]
    C2(x) -> [0.8, 0.2]
    C3(x) -> [0.4, 0.6]

#### Usando pesos uniformes, calculamos as probabilidades médias:

$$ 
p(i_{0} | X) = \frac {0.9 + 0.8 + 0.4} {3} = 0.7  
$$

<br>
<br>

$$ 
p(i_{1} | X) = \frac {0.1 + 0.2 + 0.6} {3} = 0.3  
$$

<br>
<br>

$$ 
\hat{y} = argmax_{i}[ p(i_{0} | X) , p(i_{1} | X)] = 0
$$

<br>
<br>

#### Entretanto, atribuir os pesos $\{0.1, 0.1, 0.8\}$ produziria uma previsão $ \hat{y} = 1 $:

<br>
<br>

$$ 
p(i_{0} | X) = 0.9 x 0.1 + 0.8 x 0.1 + 0.4 x 0.8 = 0.49  
$$

<br>
<br>

$$ 
p(i_{1} | X) = 0.1 x 0.1 + 0.2 x 0.1 + 0.6 x 0.8 = 0.51  
$$

<br>
<br>

$$ 
\hat{y} = argmax_{i} [ p(i_{0} | X) , p(i_{1} | X) ] = 1
$$

# Stacking Ensemble

#### O [`Stacking`](https://docs.h2o.ai/h2o/latest-stable/h2o-docs/data-science/stacked-ensembles.html) é uma técnica de aprendizado de conjunto que combina vários modelos de classificação ou regressão por meio de um meta-classificador ou meta-regressor. Os modelos de nível básico são treinados com base em um conjunto de treinamento completo; o metamodelo é treinado nas saídas do modelo de nível básico como recursos.

#### O nível básico geralmente consiste em diferentes algoritmos de aprendizado e, portanto, os conjuntos de empilhamento geralmente são heterogêneos. O algoritmo abaixo resume o empilhamento.

#### O [`Stacking`](https://medium.com/turing-talks/turing-talks-24-modelos-de-predi%C3%A7%C3%A3o-ensemble-learning-aa02ce01afda) é uma técnica comumente usada para vencer a competição de ciência de dados [`Kaggle`](https://www.kaggle.com/zaochenye/tips-for-stacking-and-blending). Por exemplo, o primeiro lugar no [desafio da Classificação de Produtos do Grupo Otto](https://www.kaggle.com/c/otto-group-product-classification-challenge) foi conquistado por um conjunto de mais de 30 modelos cuja saída foi usada como recurso para três [meta-classificadores](https://machinelearningmastery.com/meta-learning-in-machine-learning/): [`XGBoost`](https://towardsdatascience.com/exploring-xgboost-4baf9ace0cf6), [`Neural Network`](https://medium.com/datathings/meta-modelling-meta-learning-34734cd7451b) e [`Adaboost`](https://towardsdatascience.com/machine-learning-part-17-boosting-algorithms-adaboost-in-python-d00faac6c464).

#### A ideia por trás do [`stacking`](https://www.kaggle.com/zaochenye/tips-for-stacking-and-blending) se divide em $2$ partes. Queremos treinar diversos algoritmos nos nossos dados, e queremos que as visões deles sobre os algoritmos sejam complementares, ou seja, cada algoritmo aprenda algo a partir do anterior. Desse modo, o `stacking` se dá por $2$ `loops`:

 - `loop` externo: treino de cada algoritmo de modo independente.
 - `loop` interno: quebre seu dataset de treino em $10$ partes, você terá $10$ combinações de $9:1$. Treine cada algoritmo em cada uma dessas $10$ combinações. Os `outputs` de cada algoritmo no `dataset` de teste servirão de atributo para o próximo algoritmo no `loop` externo
 
De modo visual, o stacking se parece com isso:

<img src="stacking_0.png" align="left" width="30%">

#### 1) Após ter feito o `train_test_split`, quebre seu dataset de treino em $10$ partes.

<img src="stacking_1.png" align="left" width="30%"> 

#### 2) Fazendo combinações de $9:1$ partes, treine seu algoritmo (por ex, uma `Decision Tree`) nas $9$ partes e use a décima parte com pré-teste. Repita isso para todas as combinações. Isso ajuda o algoritmo a entender pontos de vista diferentes sob seu `dataset`. O nome disso é `K-Fold Cross Validation`.

<img src="stacking_2.png" align="left" width="30%">

#### 3) Após o treino do primeiro algoritmo em seu `loop` interno, faça as `predictions` no `dataset` de teste. Esse `output` será utilizado como atributo para os próximos algoritmos.

<img src="stacking_3.png" align="left" width="30%">

#### 4) Repita o passo desse `loop` externo com os próximos algoritmos, até alcançar a performance desejada.

#### Com o `Stacking` você pode criar os atributos que desejar. Por exemplo, antes de fazer um regressor para prever a quantidade de vendas de um determinado produto na próxima semana, você pode antes fazer um classificador que preve se as vendas vão aumentar ou diminuir, para então passar esse `output` como um novo atributo.

# Bagging Ensemble

#### [`Bagging`](https://medium.datadriveninvestor.com/ensemble-techniques-bagging-bootstrap-aggregating-c7a7e26bdc13) significa [`Bootstrap Aggregation`](https://corporatefinanceinstitute.com/resources/knowledge/other/bagging-bootstrap-aggregation/). Uma maneira de reduzir a variação de uma estimativa é fazer a média de várias estimativas. Por exemplo, podemos treinar M árvores diferentes em diferentes subconjuntos de dados (escolhidos aleatoriamente com substituição) e calcular o conjunto:

$$ 
f(x) = \dfrac{1}{M} \sum f_{m} (x)  
$$

#### [`Bagging`](https://subscription.packtpub.com/book/big_data_and_business_intelligence/9781789136609/5/ch05lvl1sec29/bootstrap-aggregation) usa amostragem de autoinicialização para obter os subconjuntos de dados para treinar os [`base learners`](). Para agregar os resultados dos [`base learners`](), o [`Bagging`]() usa o voto para classificação e a média para regressão.

<img src="bagging.png" align="center" width="40%">

#### Combinar `base learners` estáveis é menos vantajoso, pois o conjunto não ajudará a melhorar o desempenho da generalização. Uma classe comumente usada de algoritmos de conjuntos são [`Random Forrest`](https://machinelearningmastery.com/bagging-and-random-forest-ensemble-algorithms-for-machine-learning/).

#### A ideia por trás do [`Bagging`](https://www.quantstart.com/articles/bootstrap-aggregation-random-forests-and-boosted-trees/) é quebrar nosso `dataset` em várias partes (bags), e treinar os modelos de modo independente em cada um deles. A premissa por trás disso é que estamos utilizando modelos mais simples, sob partes dos dados, que juntos irão funcionar de modo complementar.

1) Primeiramente, nós quebramos nosso `dataset` em tantos `bags` desejarmos. O nome para isso é `bootstrapping`; <br>
2) Depois, treinamos cada sub-modelo sob cada `bag`; <br>
3) O `output` final é dado pela combinação dos `sub-outputs`. Ela pode ser feita com qualquer método de `Voting`.


### Algoritmos de bagging:

 - bagging meta-estimator
 - random forests

# Random Forrest

<img src="rf.gif" align="left" width="30%">

<br/>

#### Em [`Random Forrest`](https://machinelearningmastery.com/bagging-and-random-forest-ensemble-algorithms-for-machine-learning/), cada árvore do conjunto é construída a partir de uma amostra retirada com substituição, ou seja, uma amostra de autoinicialização do conjunto de treinamento. Além disso, em vez de usar todos os recursos, um subconjunto aleatório de recursos é selecionado, randomizando ainda mais a árvore. Como resultado, o viés da floresta aumenta um pouco, mas devido à média de árvores menos correlacionadas, sua variância diminui, resultando em um modelo geral melhor.

#### Em um algoritmo de árvores extremamente aleatórias, a aleatoriedade vai um passo além: os limites de divisão são aleatórios. Em vez de procurar o limite mais discriminativo, os limites são traçados aleatoriamente para cada recurso candidato e o melhor desses limites gerados aleatoriamente é escolhido como regra de divisão. Isso geralmente permite reduzir um pouco mais a variância do modelo, às custas de um pequeno aumento no viés.

#### A `Random Forrest` ou Floresta Aleatória, como o próprio nome indica, consiste em um grande número de árvores de decisão individuais que funcionam como um conjunto. Cada árvore individual na floresta aleatória cospe uma previsão de classe e a classe com mais votos se torna a previsão do nosso modelo.

#### Visualização de um modelo aleatório de floresta Fazendo uma previsão:

#### O conceito fundamental por trás da Floresta Aleatória é simples, mas poderoso - a sabedoria das multidões. Em termos de ciência de dados, a razão pela qual o modelo aleatório de floresta funciona tão bem é: Um grande número de modelos (árvores) relativamente não correlacionados que opera como um comitê superará qualquer um dos modelos constituintes individuais.

#### A baixa correlação entre os modelos é a chave. Assim como os investimentos com baixas correlações, como ações e títulos, se juntam para formar um portfólio maior que a soma de suas partes, os modelos não correlacionados podem produzir previsões de conjuntos mais precisas do que as previsões individuais. 

#### A razão para esse maravilhoso efeito é que as árvores se protegem de seus erros individuais (desde que nem sempre erram na mesma direção). Enquanto algumas árvores podem estar erradas, muitas outras estão certas, então, como um grupo, as árvores podem se mover na direção correta. Portanto, os pré-requisitos para um desempenho aleatório da floresta aleatória são:

- É necessário que haja algum sinal real em nossos recursos, para que os modelos criados com esses recursos sejam melhores do que a adivinhação aleatória.

- As previsões e portanto, os erros, feitos pelas árvores individuais precisam ter baixas correlações entre si.

### Garantir que os modelos se diversifiquem

#### Então, como a floresta aleatória garante que o comportamento de cada árvore individual não esteja muito correlacionado com o comportamento de qualquer outra árvore no modelo? Ele usa os dois métodos a seguir:

- Bagging (Agregação de Bootstrap): As árvores de decisãao são muito sensíveis aos dados nos quais são treinadas, pequenas alterações no conjunto de treinamento podem resultar em estruturas de árvore significativamente diferentes. A floresta aleatória tira vantagem disso, permitindo que cada árvore individual faça uma amostragem aleatória do conjunto de dados com substituição, resultando em árvores diferentes. Esse processo é conhecido como `Bagging`.

#### Observe que, com o `Bagging`, não estamos agrupando os dados de treinamento em blocos menores e treinando cada árvore em um bloco diferente. Em vez disso, se tivermos uma amostra do tamanho $N$, ainda estaremos alimentando cada árvore com um conjunto de treinamento do tamanho $N$, a menos que especificado de outra forma. Mas, em vez dos dados de treinamento originais, coletamos uma amostra aleatória do tamanho $N$ com substituição. Por exemplo, se nossos dados de treinamento foram $[1, 2, 3, 4, 5, 6]$, podemos fornecer a uma de nossas árvores a seguinte lista $[1, 2, 2, 3, 6, 6]$. Observe que ambas as listas têm o comprimento $6$ e que $2$ e $6$ são repetidos nos dados de treinamento selecionados aleatoriamente que fornecemos à nossa árvore (porque fazemos a amostragem com substituição).

- Feature Randomness (Aleatoriedade de Atrinutos): Em uma árvore de decisão normal, quando é hora de dividir um nó, consideramos todos os recursos possíveis e escolhemos o que produz a maior separação entre as observações no nó esquerdo e as do nó direito. Por outro lado, cada árvore em uma floresta aleatória pode escolher apenas um subconjunto aleatório de recursos. Isso força ainda mais variação entre as árvores no modelo e, finalmente, resulta em menor correlação entre as árvores e mais diversificação.

## A Divisão

#### A divisão de nós em um modelo de floresta aleatória é baseada em um subconjunto aleatório de recursos para cada árvore:

A Árvore $2$, por outro lado, só pode ver os Recursos $1$ e $3$, para poder escolher o Recurso $1$.

#### Portanto, em nossa floresta aleatória, acabamos com árvores que não são apenas treinadas em diferentes conjuntos de dados, graças ao `bagging`, mas também usam recursos diferentes para tomar decisões.

### Revisando

#### A Floresta Aleatória é um algoritmo de classificação que consiste em muitas árvores de decisão. Ele usa `bagging` e aleatoriedade de atributos ao construir cada árvore individual para tentar criar uma floresta não correlacionada de árvores cuja previsão pelo comitê é mais precisa do que a de qualquer árvore individual.

#### Precisamos de atributos que tenham pelo menos algum poder preditivo. Afinal, se colocarmos lixo, então vamos tirar o lixo. As árvores da floresta e, mais importante, suas previsões precisam ser não correlacionadas, ou pelo menos ter baixas correlações entre si. Enquanto o próprio algoritmo via aleatoriedade de atributos tenta projetar essas baixas correlações para nós, os recursos que selecionamos e os hiperparâmetros que escolhemos também afetarão as correlações finais.

## O Algoritmo

#### Para constuir uma Random Forest de N árvores, seguimos os seguintes passos:

* Para cada $k = 1, \dots, N$:

    * Gerar um `dataset` `boostrap` $X_{k}$.
    * Construir uma árvore de decisão $b_{k}$ em $X_{k}$:

        * Escolher o melhor atributo de acordo com algum critério (ex: gini). Quebrar a amostra por esse atributo para uma nova camada na árvore. 
        * Construir a árvore até que suas folhas não contenham mais que $n_\text{min}$ instâncias ou até uma certa profundidade seja atingida (parada da recursão).
        * Para cada quebra, escolher randomicamente  $m$ features de $d$ originais e então procurar pelo próximo melhor. 

## O classificador final é definido por:
$$
a(x) = \frac{1}{N}\sum_{k = 1}^{N} b_k(x)
$$

#### Nós utilizamos majority voting para classificação e média/mediana para regressão.

#### Para problemas de classificação, é recomendado utilizar $m = \sqrt{d}$. Para regressões, nós utilizamos $m = \frac{d}{3}$, onde $\large d$ é o número de features. É recomendado que cada árvore seja construída até que todas suas folhas contenham somente $n_\text{min} = 1$ examples para classificação e $n_\text{min} = 5$ exemplos para regressão.

### Prática: Árvores vs Bagging vs RFs:

#### Vamos começar observando o comportamento desses $3$ algoritmos num cenário de regressão.

```python
# importando as bibliotecas
# Disable warnings
import warnings
import numpy as np
warnings.filterwarnings('ignore')

%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = 10, 6
%config InlineBackend.figure_format = 'retina' 

import seaborn as sns
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
n_train = 150        
n_test = 1000       
noise = 0.1

# Generate data
def f(x):
    x = x.ravel()
    return np.exp(- x ** 2) + 1.5 * np.exp(- (x - 2) ** 2)

def generate(n_samples, noise):
    X = np.random.rand(n_samples) * 10 - 5
    X = np.sort(X).ravel()
    y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2) 
    + np.random.normal(0.0, 
                       noise, 
                       n_samples
                      )
    X = X.reshape((n_samples, 1))

    return X, y

X_train, y_train = generate(n_samples = n_train, 
                            noise = noise 
                           )
X_test, y_test = generate(n_samples = n_test, 
                          noise = noise
                         )
# One decision tree regressor
dtree = DecisionTreeRegressor().fit(X_train, 
                                    y_train
                                   )
d_predict = dtree.predict(X_test)

plt.figure(figsize = (10, 6))
plt.plot(X_test, f(X_test), "b")
plt.scatter(X_train, y_train, c = "b", s = 20)
plt.plot(X_test, d_predict, "g", lw = 2)
plt.xlim([-5, 5])
plt.title("Decision tree, MSE = %.2f" % np.sum((y_test - d_predict) ** 2))

# Bagging with a decision tree regressor
bdt = BaggingRegressor(DecisionTreeRegressor()).fit(X_train, 
                                                    y_train
                                                   )
bdt_predict = bdt.predict(X_test)

plt.figure(figsize = (10, 6))
plt.plot(X_test, f(X_test), "b")
plt.scatter(X_train, y_train, c = "b", s = 20)
plt.plot(X_test, bdt_predict, "y", lw = 2)
plt.xlim([- 5, 5])
plt.title("Bagging for decision trees, MSE = %.2f" % np.sum((y_test - bdt_predict) ** 2));

# Random Forest
rf = RandomForestRegressor(n_estimators = 10).fit(X_train, 
                                                y_train
                                               )
rf_predict = rf.predict(X_test)

plt.figure(figsize = (10, 6))
plt.plot(X_test, f(X_test), "b")
plt.scatter(X_train, y_train, c = "b", s = 20)
plt.plot(X_test, rf_predict, "r", lw = 2)
plt.xlim([-5, 5])
plt.title("Random forest, MSE = %.2f" % np.sum((y_test - rf_predict) ** 2));
```

Agora, vamos observar o comportamento desses algoritmos num cenário de classificação:

```python
np.random.seed(42)
X, y = make_circles(n_samples = 500, 
                    factor = 0.1, 
                    noise = 0.35, 
                    random_state = 42
                   )
X_train_circles, X_test_circles, y_train_circles, y_test_circles =  train_test_split(X, y, test_size = 0.2)

dtree = DecisionTreeClassifier(random_state = 42)
dtree.fit(X_train_circles, y_train_circles)

x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, 
                       x_range
                      )
y_hat = dtree.predict(np.c_[xx1.ravel(), 
                            xx2.ravel()]
                     )
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, 
             xx2, 
             y_hat, 
             alpha = 0.2
            )
plt.scatter(X[:,0], 
            X[:,1], 
            c = y, 
            cmap = 'viridis', 
            alpha = 0.7
           )
plt.title("Decision tree")
plt.show()

b_dtree = BaggingClassifier(DecisionTreeClassifier(), 
                            n_estimators = 300, 
                            random_state = 42
                           )
b_dtree.fit(X_train_circles, 
            y_train_circles
           )
x_range = np.linspace(X.min(), 
                      X.max(), 
                      100
                     )
xx1, xx2 = np.meshgrid(x_range, 
                       x_range
                      )
y_hat = b_dtree.predict(np.c_[xx1.ravel(), 
                              xx2.ravel()]
                       )
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, 
             xx2, 
             y_hat, 
             alpha = 0.2
            )
plt.scatter(X[:, 0], 
            X[:, 1], 
            c = y, 
            cmap = 'viridis', 
            alpha = 0.7
           )
plt.title("Bagging (decision trees)")
plt.show()

rf = RandomForestClassifier(n_estimators = 300, 
                            random_state = 42
                           )
rf.fit(X_train_circles, 
       y_train_circles
      )
x_range = np.linspace(X.min(), 
                      X.max(), 
                      100
                     )
xx1, xx2 = np.meshgrid(x_range, 
                       x_range
                      )
y_hat = rf.predict(np.c_[xx1.ravel(), 
                         xx2.ravel()]
                  )
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, 
             xx2, 
             y_hat, 
             alpha = 0.2
            )
plt.scatter(X[:, 0], 
            X[:, 1], 
            c = y, 
            cmap = 'viridis', 
            alpha = 0.7
           )
plt.title("Random forest")
plt.show()
```

#### Nas figuras acima, podemos ver que a superfície de decisão da árvore de decisão é muito irregular e anguloso, o que pode levar a situações de `overfitting`, ou seja, baixa habilidade de generalização. Nos exemplos de `bagging` e `RF`, vemos que a superfície de separação é muito mais suave e generalista.
